```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

start_value_chur = 0.279718
end_value_chur = 0.28320

start_value_depleted_mantle = 0.279718
end_value_depleted_mantle = 0.28420

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

time_points = [0, 4.5]

def calculate_isotope_ratio(time, start_value, end_value):
    if time[-1] == time[0]:
        return [start_value] * len(time)
    slope = (end_value - start_value) / (time[-1] - time[0])
    return [start_value + slope * (t - time[0]) for t in time]

line_chur, = ax.plot([], [], label='CHUR Evolution')
line_depleted_mantle, = ax.plot([], [], label='Depleted Mantle')

ax.set_xlim(0, 4.5)
ax.set_ylim(0.2795, 0.2845)
ax.set_xlabel('Time (Ga)')
ax.set_ylabel('176/177 Hf Isotope Ratio')
ax.legend()

def update(frame):
    time = [4.5 - frame / 10, 0]
    ratio_chur = calculate_isotope_ratio(time, start_value_chur, end_value_chur)
    ratio_depleted_mantle = calculate_isotope_ratio(time, start_value_depleted_mantle, end_value_depleted_mantle)

    line_chur.set_data(time, ratio_chur)
    line_depleted_mantle.set_data(time, ratio_depleted_mantle)

    return line_chur, line_depleted_mantle

animation = FuncAnimation(fig, update, frames=range(0, 51), interval=100, blit=True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


half_life_U238 = 4.468e9  
half_life_Th232 = 1.405e10  
ratio_Hf176_Hf177_present = 0.282793 
ratio_U238_Pb206_present = 0.05527  

def hafnium_isotope_ratio(t):
    ratio_U238_Pb206_initial = 1 - ratio_Hf176_Hf177_present 
    decay_U238_to_Pb206 = np.exp(-np.log(2) * t / half_life_U238) 
    ratio_U238_Pb206 = ratio_U238_Pb206_initial * decay_U238_to_Pb206 
    ratio_Hf176_Hf177 = 1 - ratio_U238_Pb206  
    return ratio_Hf176_Hf177

t = np.linspace(0, 4.5e9, 1000) 

ratio_Hf176_Hf177_evolution = hafnium_isotope_ratio(t)

plt.plot(t, ratio_Hf176_Hf177_evolution, label="Hafnium-Isotopenverhältnis")
plt.xlabel('Zeit (Jahre)')
plt.ylabel('Verhältnis von Hafnium-176 zu Hafnium-177')
plt.title('Entwicklung des Hafnium-Isotopenverhältnisses über Milliarden von Jahren')
plt.legend()
plt.show()
```
