```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

def generate_info_list(url):
    
    driver = webdriver.Chrome()

    # Load the page using the driver
    driver.get(url)

    # Wait for the page to load completely
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, "title-text")))

    # Get the page source and parse it with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Find the elements using BeautifulSoup
    titles = soup.find_all("a", class_="title-text")
    givennames = soup.find_all("a", class_="given-name")
    names = soup.find_all("a", class_="text surname")
    abstracts = soup.find_all("a", class_="sp0005")

    # Store the data in a dictionary
    attributes = 'title', 'givenname', 'name', 'abstract',
    data2 = {key: [] for key in attributes}

    # Extract the text from the elements and store it in the dictionary
    for title in titles:
        data2['title'].append(title.text)
    for givenname in givennames:
        data2['givenname'].append(givenname.text)
    for name in names:
        data2['name'].append(name.text)
    for abstract in abstracts:
        data2['abstract'].append(abstract.text)

    # Create a DataFrame from the dictionary
    df2 = pd.DataFrame(data2)

    # Close the driver
    driver.quit()

    return df2

# Read the Volumes.csv file into a Pandas DataFrame
df = pd.read_csv('Volumes.csv')

# Extract the links from the hyperlink column
links = df['hyperlink'].tolist()

# Initialize an empty DataFrame to store the results
all_df2 = pd.DataFrame()

# Loop through the links and generate the info list for each one
for link in links:
    df2 = generate_info_list(link)
    all_df2 = pd.concat([all_df2, df2])
    print(all_df2.title)
    time.sleep(1) # Add a delay between requests to avoid overwhelming the server

all_df2.to_csv('Semesterprojekt.csv')
```
